```python
from prophet import Prophet
import matplotlib.pyplot as plt
```

```python
def predict_aqi(city, input_date, df):
    # Filter data for the specified city
    city_df = df[df['City'] == city].copy()

    #city_df['ds'] = pd.to_datetime(city_df['ds'])

    # Rename columns to 'ds' and 'y' as required by Prophet
    city_df = city_df.rename(columns={'ds': 'ds', 'y': 'y'})

    model = Prophet()
    model.fit(city_df)

    # Create a dataframe with the input date
    future = pd.DataFrame({'ds': [input_date]})

    # Make predictions for the input date
    forecast = model.predict(future)


    predicted_aqi = forecast.loc[0, 'yhat']

    return predicted_aqi
```

```python
historical_data = pd.read_csv('output.csv')

# Get user input for city and date
user_city = input("Enter the city: ")
user_date = input("Enter the date (YYYY-MM-DD): ")

# Call the predict_aqi function
pred = predict_aqi(user_city, user_date, historical_data)

print(f"The expected AQI for {user_city} on {user_date} is {pred:.2f}")

if pred > 0 and pred <= 50:
    print('The AQI is good')
elif pred > 51 and pred <=100:
    print('The AQI is moderate however, there may be a risk for some people, particularly those who are unusually sensitive to air pollution.')
elif pred > 101 and pred <=150:
    print('The AQI is unhealthy for sensitive groups the general public is less likely to be affected.')
elif pred>151 and pred <= 200:
    print('The AQI is unhealthy, members of sensitive groups may experience more serious health effects.')
elif pred>201 and pred<=300:
    print('The AQI is very unhealthy. The risk of health effects is increased for everyone.')
elif pred>301 and pred<=500:
    print('The AQI is hazardous. People with heart or lung disease, older adults, and children should avoid all outdoor activity.')
```
